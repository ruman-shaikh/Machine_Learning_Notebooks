```python
#Importing Modules

import numpy as np
import pandas as pd
```

```python
#Understanding the Dataset

file = pd.read_csv('telescope_data.csv')
print(file.dtypes)
file.fillna(0) 
file.head()
#file.tail()
```

```python
print("Output classes: ", file['class'].unique())
```

```python
#Data Encoding

from sklearn import preprocessing

data_encoder = preprocessing.LabelEncoder()
file['class'] = data_encoder.fit_transform(file['class'])

#Testing Encoding
print(file.dtypes)
file.head()
#file.tail()
```

```python
X = file.iloc[:,1:11].copy() #copying columns 0 to 9
print(file.shape, X.shape)
X.head()
```

```python
Y = file[['class']].copy()
print(Y.shape)
Y.head()
```

```python
#Train/Test Split
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=15)
print(X_train.shape, X_test.shape)
```

```python
#Machine Lerning

#Support Vector Machine(SVM)
from sklearn import svm
from sklearn.metrics import accuracy_score

clf = svm.SVC(gamma='scale')
clf.fit(X_train, Y_train.values.ravel())

pred = clf.predict(X_test)
print('Accuracy using Support Vector Machine(SVM): ', 100 * accuracy_score(Y_test.values.ravel(), pred))
```

```python
#Multilayer Perceptron
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score

clf = MLPClassifier(activation='logistic', solver='adam', alpha=1e-5, hidden_layer_sizes=(7, 5), learning_rate='adaptive', random_state=1)
clf.fit(X_train, Y_train.values.ravel())

pred = clf.predict(X_test)
print('Accuracy using Multilayer Perceptron: ', 100 * accuracy_score(Y_test.values.ravel(), pred))
```

```python
#Gaussian Naive Bayes
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

clf = GaussianNB()
clf.fit(X_train, Y_train.values.ravel())

pred = clf.predict(X_test)
print('Accuracy using Gaussian Naive Bayes: ', 100 * accuracy_score(Y_test.values.ravel(), pred))
```
